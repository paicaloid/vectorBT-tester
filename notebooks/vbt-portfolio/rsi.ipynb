```python
rsi = vbt.pandas_ta("RSI").run(df["close"], length=14)
df['entries'] = rsi.rsi_crossed_below(40).shift(1).fillna(False)
df['exits'] = rsi.rsi_crossed_above(60).shift(1).fillna(False)
```

```python
start_date = "2018-01-01"
end_date = "2023-03-31"
df_adj = df.loc[(df.index >= start_date) & (df.index <= end_date)]
```

```python
pf_vbt = vbt.Portfolio.from_signals(df_adj["open"], df_adj["entries"], df_adj["exits"], init_cash=200, size=1)
```
