```python
import vectorbtpro as vbt
import numpy as np
import pandas as pd

from indicators.adx_dx import ADX
vbt.settings.set_theme("dark")
```

```python
data = vbt.BinanceData.fetch('BTCUSDT', start="2021-01")
```

```python
data.plot().show() 
```

```python
adx, plusDI, minusDI = ADX(data.high, data.low, data.close, period=14)
```

```python
adx_level = 25
entries_long = (plusDI > minusDI) & (plusDI > adx_level)
exits_long = (plusDI < minusDI) | (plusDI < adx_level)
```

```python
start_date = "2022-01-01"
stop_date = "2023-04-04"
mask = (data.close.index >= start_date) & (data.close.index <= stop_date)
```

```python
pf = vbt.Portfolio.from_signals(
    # close=data.data["BTCUSDT"].loc[mask],
    # price=data.close[mask],
    close=data.close[mask],
    open=data.open[mask],
    high=data.high[mask],
    low=data.low[mask],
    entries=entries_long[mask],
    exits=exits_long[mask],
    size=1,
    init_cash=50000,
    from_ago=1,
)
```

```python
pf.plot().show()
```
